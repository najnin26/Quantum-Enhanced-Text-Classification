```python
# ========================== INSTALLS & IMPORTS ==========================

# COMMENT OUT IF ALREADY INSTALLED ON YOUR ENV
# !pip install pandas scikit-learn torch transformers sentencepiece nltk pennylane matplotlib seaborn
!pip install pennylane -q

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset, RandomSampler, SequentialSampler
import time
import random
import re
import warnings
import nltk
import pennylane as qml
from pennylane import numpy as pnp

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    f1_score,
    precision_score,
    recall_score,
    roc_curve,
    roc_auc_score,
)

from transformers import (
    AutoTokenizer,
    AutoModel,
    get_linear_schedule_with_warmup,
)
from torch.optim import AdamW
import matplotlib.pyplot as plt
import seaborn as sns

nltk.download("stopwords", quiet=True)
warnings.filterwarnings("ignore")


# ========================== QUANTUM CIRCUIT DEFINITION ==========================

# Create a quantum node (circuit)
dev = qml.device("default.qubit", wires=4)  # 4-qubit quantum circuit for better expressivity

@qml.qnode(dev, interface='torch')  # Important: specify torch interface for gradient flow
def quantum_circuit(weights, inputs):
    """
    Quantum circuit for enhanced feature processing
    Args:
        weights: trainable quantum parameters (shape: 4)
        inputs: classical input features (shape: 4)
    Returns:
        expectation values of PauliZ on all 4 wires
    """
    # Initial Hadamard gates to create superposition on all qubits
    for i in range(4):
        qml.Hadamard(wires=i)

    # Encode classical inputs using rotation gates
    for i in range(4):
        qml.RX(inputs[i], wires=i)
        qml.RY(inputs[i] * 0.5, wires=i)  # Additional encoding for richer representation

    # Entanglement layers (ring topology)
    for i in range(3):
        qml.CNOT(wires=[i, i+1])
    qml.CNOT(wires=[3, 0])  # Close the ring

    # Second entanglement layer for better information flow
    for i in range(3):
        qml.CZ(wires=[i, i+1])
    
    # Trainable quantum weights (rotation gates)
    for i in range(4):
        qml.RZ(weights[i], wires=i)
        qml.RX(weights[i] * 0.3, wires=i)  # Additional trainable rotation

    # Third entanglement layer
    for i in range(4):
        for j in range(i+1, 4):
            if j - i <= 2:  # Connect nearby qubits
                qml.CNOT(wires=[i, j])

    # Measure expectation values for all qubits
    return [qml.expval(qml.PauliZ(i)) for i in range(4)]


# ========================== DATA & PATH ==========================

# Change this to your actual CSV path
CSV_PATH = "/kaggle/input/datasets/najninsultanashirin/spam-dataset-qtl/spamdata_v2 (1).csv"

df = pd.read_csv(CSV_PATH)
print("Initial data shape:", df.shape)


# ========================== TEXT PREPROCESSING ==========================

def preprocess_text(text):
    text = re.sub(r'\\s+', ' ', str(text))  # Remove extra spaces
    text = re.sub(r'http\\S+|www\\S+|https\\S+', '', text)  # Remove links
    text = re.sub(r'@\w+', '', text)  # Remove mentions
    text = re.sub(r'#\w+', '', text)  # Remove hashtags
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)  # Remove punct (keep alphanum)
    text = text.lower().strip()
    return text


df["cleaned_text"] = df["text"].apply(preprocess_text)

# Filter out special tokens
df = df[df["cleaned_text"] != "[deleted]"]
df = df[df["cleaned_text"] != "[removed]"]


# Get NLTK stopwords
stop_words = set(nltk.corpus.stopwords.words("english"))

def clean_statement(statement):
    statement = statement.lower()
    statement = re.sub(r'[^a-zA-Z0-9\s]', ' ', statement)
    statement = re.sub(r'\d+', '', statement)
    words = statement.split()
    words = [w for w in words if w not in stop_words]
    return ' '.join(words)


df["cleaned_text"] = df["cleaned_text"].apply(clean_statement)

# Label encoding
label_encoder = LabelEncoder()
df["label"] = label_encoder.fit_transform(df["label"])

print("Classes after label encoding:", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))


# ========================== BALANCE MINOR/MAXOR CLASSES ==========================

class_counts = df["label"].value_counts()
print("\nClass counts before balancing:")
print(class_counts)

majority_class = class_counts.idxmax()
minority_class = class_counts.idxmin()

df_majority = df[df["label"] == majority_class]
df_minority = df[df["label"] == minority_class]

df_majority_downsampled = df_majority.sample(len(df_minority), random_state=42)
df_balanced = pd.concat([df_majority_downsampled, df_minority]).reset_index(drop=True)
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print("\nClass counts after balancing:")
print(df_balanced["label"].value_counts())

# Optional: inspect class distribution plot
sns.countplot(x="label", data=df_balanced)
plt.title("Balanced Class Distribution")
plt.show()


# ========================== TOKENIZER & ENCODING ==========================

MODEL_NAME = "microsoft/deberta-v3-base"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

MAX_LEN = 128
texts = df_balanced["cleaned_text"].tolist()
labels = df_balanced["label"].values

tokens = tokenizer(
    texts,
    max_length=MAX_LEN,
    padding="max_length",
    truncation=True,
    return_tensors="pt",
)

input_ids  = tokens["input_ids"]   # shape (N, MAX_LEN)
attention_masks = tokens["attention_mask"]
labels     = torch.tensor(labels)


# ========================== DEVICE SETUP ==========================

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


# ========================== SIMPLE EVALUATION FUNCTION ==========================

def flat_accuracy(preds, labels):
    pred_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()
    return np.sum(pred_flat == labels_flat) / len(labels_flat)


# ========================== QUANTUM ENHANCED MODEL: QBiLSTM with Quantum Layer ==========================

class QBiLSTM(nn.Module):
    def __init__(self, num_classes=2, freeze_bert=False, use_quantum=True):
        super(QBiLSTM, self).__init__()
        
        self.use_quantum = use_quantum
        self.bert = AutoModel.from_pretrained(MODEL_NAME)

        if freeze_bert:
            for p in self.bert.parameters():
                p.requires_grad = False

        self.dropout = nn.Dropout(0.3)

        # BiLSTM layer (bidirectional)
        self.bilstm = nn.LSTM(
            input_size=self.bert.config.hidden_size,  # deberta-v3-base has 1024 hidden size
            hidden_size=128,
            batch_first=True,
            bidirectional=True,
        )
        
        # If using quantum layer, we'll project features to quantum-compatible size
        if self.use_quantum:
            # Project from 256 (BiLSTM output) to 4 (quantum input features)
            self.quantum_projection = nn.Linear(256, 4)
            self.quantum_batch_norm = nn.BatchNorm1d(4)  # Normalize quantum features
            # Trainable quantum weights (4 parameters for the 4-qubit circuit)
            self.quantum_weights = nn.Parameter(torch.randn(4, dtype=torch.float32) * 0.1)
            # Quantum feature processing layer
            self.quantum_feature_processor = nn.Linear(4, 16)  # Expand quantum features
            self.quantum_dropout = nn.Dropout(0.2)
        
        # Final classification layer
        # If using quantum, input size is 256 (classical) + 16 (quantum features)
        # If not using quantum, input size is 256
        if self.use_quantum:
            self.fc = nn.Sequential(
                nn.Linear(256 + 16, 128),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(128, 64),
                nn.ReLU(),
                nn.Dropout(0.2),
                nn.Linear(64, num_classes)
            )
        else:
            self.fc = nn.Sequential(
                nn.Linear(256, 128),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(128, num_classes)
            )

    def quantum_layer(self, features):
        """
        Process features through quantum circuit with batch processing optimization
        Args:
            features: classical features (batch_size, 4) - projected from BiLSTM output
        Returns:
            quantum_output: quantum processed features (batch_size, 16) - expanded features
        """
        batch_size = features.shape[0]
        quantum_outputs = []
        
        # Process each sample in the batch
        for i in range(batch_size):
            # Get the features for quantum processing
            quantum_input = features[i].detach().cpu().numpy()
            
            # Get quantum weights
            weights = self.quantum_weights.detach().cpu().numpy()
            
            # Execute quantum circuit
            try:
                q_output = quantum_circuit(weights, quantum_input)
                # q_output is a list of 4 expectation values
                q_output_tensor = torch.tensor(q_output, dtype=torch.float32).to(device)
            except Exception as e:
                # Fallback to classical processing if quantum fails
                print(f"Quantum processing error: {e}, using classical fallback")
                q_output_tensor = features[i]
            
            quantum_outputs.append(q_output_tensor)
        
        # Stack all quantum outputs
        quantum_outputs = torch.stack(quantum_outputs)  # (batch_size, 4)
        
        # Normalize quantum outputs
        quantum_outputs = self.quantum_batch_norm(quantum_outputs)
        
        # Process quantum features through classical layer to expand dimensionality
        quantum_outputs = self.quantum_feature_processor(quantum_outputs)  # (batch_size, 16)
        quantum_outputs = self.quantum_dropout(quantum_outputs)
        
        return quantum_outputs

    def forward(self, input_ids, attention_mask):
        # BERT encoding
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        encoder_output = outputs.last_hidden_state
        
        # Ensure float32 for LSTM
        encoder_output = encoder_output.to(torch.float32)
        
        # Apply dropout and BiLSTM
        lstm_out, (hidden, cell) = self.bilstm(self.dropout(encoder_output))
        
        # Get last hidden state (batch_size, 256) because bidirectional (128*2)
        last_hidden = lstm_out[:, -1, :]
        
        if self.use_quantum:
            # Project BiLSTM features to quantum-compatible size (4 features)
            quantum_features = self.quantum_projection(last_hidden)
            
            # Apply quantum processing
            try:
                quantum_output = self.quantum_layer(quantum_features)
                
                # Concatenate classical and quantum features
                combined_features = torch.cat([last_hidden, quantum_output], dim=1)
                
                # Final classification
                logits = self.fc(combined_features)
            except Exception as e:
                print(f"Error in quantum layer: {e}, falling back to classical")
                logits = self.fc(last_hidden)
        else:
            # Classical path only
            logits = self.fc(last_hidden)
        
        return logits


# ========================== TRAINING FUNCTION WITH SEEDS ==========================

SEEDS = [42, 123, 256, 512, 999]

all_results = []
classical_results = []  # For comparison


def train_model_with_seed(seed, input_ids, attention_masks, labels, batch_size=32, lr=2e-5, epochs=5, use_quantum=True):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

    print(f"\n{'='*80}")
    print(f"TRAINING WITH SEED = {seed}")
    print(f"Optimizer: AdamW, Learning Rate: {lr}, Batch Size: {batch_size}")
    print(f"Quantum Layer: {'ENABLED' if use_quantum else 'DISABLED'}")
    print(f"{'='*80}")

    num_classes = len(torch.unique(labels))
    
    # Stratified 80/20 split
    train_inputs, val_inputs, train_labels, val_labels = train_test_split(
        input_ids,
        labels,
        test_size=0.2,
        random_state=seed,
        stratify=labels,
    )
    train_masks, val_masks, _, _ = train_test_split(
        attention_masks,
        labels,
        test_size=0.2,
        random_state=seed,
        stratify=labels,
    )

    # Datasets
    train_dataset = TensorDataset(train_inputs, train_masks, train_labels)
    val_dataset   = TensorDataset(val_inputs,   val_masks,   val_labels)

    train_loader = DataLoader(train_dataset, sampler=RandomSampler(train_dataset), batch_size=batch_size)
    val_loader   = DataLoader(val_dataset,   sampler=SequentialSampler(val_dataset), batch_size=batch_size)

    # Model
    model = QBiLSTM(num_classes=num_classes, use_quantum=use_quantum).to(device)
    model = model.to(torch.float32)
    
    # Count total parameters
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"\nModel parameters: Total={total_params:,}, Trainable={trainable_params:,}")

    # Optimizer & scheduler
    optimizer = AdamW(model.parameters(), lr=lr)
    total_steps = len(train_loader) * epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps,
    )

    # Lists to store metrics
    train_losses = []
    train_accs   = []
    val_losses   = []
    val_accs     = []

    best_val_acc = 0.0
    best_model_state = None

    # For final report
    all_preds = []
    all_trues = []

    for epoch in range(epochs):
        start_time = time.time()

        # --- Training phase ---
        model.train()
        total_train_loss = 0
        total_train_acc  = 0

        for batch_idx, batch in enumerate(train_loader):
            b_input_ids, b_input_mask, b_labels = batch
            b_input_ids    = b_input_ids.to(device)
            b_input_mask   = b_input_mask.to(device)
            b_labels       = b_labels.to(device)

            optimizer.zero_grad()

            logits = model(b_input_ids, b_input_mask)

            loss_fn = nn.CrossEntropyLoss()
            loss = loss_fn(logits, b_labels)

            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

            optimizer.step()
            scheduler.step()

            total_train_loss += loss.item()

            # For accuracy
            preds = F.softmax(logits, dim=1).detach().cpu().numpy()
            trues = b_labels.cpu().numpy()
            total_train_acc += flat_accuracy(preds, trues)

        avg_train_loss = total_train_loss / len(train_loader)
        avg_train_acc  = total_train_acc  / len(train_loader)
        train_losses.append(avg_train_loss)
        train_accs.append(avg_train_acc)

        # --- Validation phase ---
        model.eval()
        total_val_loss = 0
        total_val_acc  = 0

        val_preds = []
        val_trues = []

        with torch.no_grad():
            for batch in val_loader:
                b_input_ids, b_input_mask, b_labels = batch
                b_input_ids    = b_input_ids.to(device)
                b_input_mask   = b_input_mask.to(device)
                b_labels       = b_labels.to(device)

                logits = model(b_input_ids, b_input_mask)
                loss = nn.CrossEntropyLoss()(logits, b_labels)

                total_val_loss += loss.item()

                preds = F.softmax(logits, dim=1).detach().cpu().numpy()
                trues = b_labels.cpu().numpy()
                total_val_acc += flat_accuracy(preds, trues)

                val_preds.extend(np.argmax(preds, axis=1).flatten())
                val_trues.extend(trues.flatten())

        avg_val_loss = total_val_loss / len(val_loader)
        avg_val_acc  = total_val_acc  / len(val_loader)
        val_losses.append(avg_val_loss)
        val_accs.append(avg_val_acc)

        epoch_time = time.time() - start_time

        print(
            f"Epoch {epoch+1:2d}/{epochs} | "
            f"Train Loss: {avg_train_loss:6.4f} | Train Acc: {avg_train_acc:.4f} | "
            f"Val Loss: {avg_val_loss:6.4f} | Val Acc: {avg_val_acc:.4f} | "
            f"Time: {epoch_time:5.2f}s"
        )

        # --- Save best model ---
        if avg_val_acc > best_val_acc:
            best_val_acc = avg_val_acc
            best_model_state = model.state_dict().copy()

        all_preds = val_preds
        all_trues = val_trues

    # Final classification report
    print(f"\nFinal Classification Report for Seed {seed}:")
    print(classification_report(all_trues, all_preds, target_names=["Not Spam", "Spam"]))

    # Confusion matrix
    cm = confusion_matrix(all_trues, all_preds)

    # Extra metrics
    f1      = f1_score(all_trues, all_preds, average="weighted")
    prec    = precision_score(all_trues, all_preds, average="weighted")
    recall  = recall_score(all_trues, all_preds, average="weighted")

    results = {
        "seed": seed,
        "use_quantum": use_quantum,
        "best_val_accuracy": best_val_acc,
        "final_train_accuracy": train_accs[-1],
        "final_val_accuracy":   val_accs[-1],
        "final_train_loss":     train_losses[-1],
        "final_val_loss":       val_losses[-1],
        "f1_score": f1,
        "precision": prec,
        "recall": recall,
        "confusion_matrix": cm,
        "train_accuracies": train_accs,
        "val_accuracies": val_accs,
        "train_losses": train_losses,
        "val_losses": val_losses,
        "predictions": np.array(all_preds),
        "true_labels": np.array(all_trues),
    }

    return results


# ========================== RUN 5 SEEDS WITH QUANTUM LAYER ==========================

print("\n" + "="*80)
print("STARTING TRAINING WITH SPAM DETECTION - QUANTUM ENHANCED MODEL")
print(f"Optimizer: AdamW, Learning Rate: 2e-5, Batch Size: 32")
print(f"Quantum Circuit: 4-qubit with entanglement and multiple layers")
print("="*80)

print("\n>>> TRAINING QUANTUM ENHANCED MODEL (with quantum layer)")
for seed in SEEDS:
    results = train_model_with_seed(
        seed=seed,
        input_ids=input_ids,
        attention_masks=attention_masks,
        labels=labels,
        batch_size=32,
        lr=2e-5,
        epochs=5,
        use_quantum=True,  # Enable quantum layer
    )
    all_results.append(results)


# ========================== RUN 5 SEEDS WITHOUT QUANTUM LAYER (FOR COMPARISON) ==========================

# print("\n" + "="*80)
# print("TRAINING CLASSICAL MODEL (without quantum layer) FOR COMPARISON")
# print("="*80)

# print("\n>>> TRAINING CLASSICAL MODEL (no quantum layer)")
# for seed in SEEDS[:2]:  # Run only 2 seeds for comparison to save time (can change to all 5)
#     classical_results.append(train_model_with_seed(
#         seed=seed,
#         input_ids=input_ids,
#         attention_masks=attention_masks,
#         labels=labels,
#         batch_size=32,
#         lr=2e-5,
#         epochs=5,
#         use_quantum=False,  # Disable quantum layer
#     ))


# ========================== SUMMARY TABLE ==========================

print("\n" + "="*80)
print("COMPREHENSIVE RESULTS FOR ALL 5 SEEDS (Quantum Enhanced)")
print("="*80)

summary_data = []
for results in all_results:
    summary_data.append({
        "Seed": results["seed"],
        "Best Val Acc":  f"{results['best_val_accuracy']:.4f}",
        "Final Val Acc": f"{results['final_val_accuracy']:.4f}",
        "Final Train Acc": f"{results['final_train_accuracy']:.4f}",
        "F1-Score": f"{results['f1_score']:.4f}",
        "Precision": f"{results['precision']:.4f}",
        "Recall": f"{results['recall']:.4f}",
        "Final Val Loss": f"{results['final_val_loss']:.4f}",
    })

summary_df = pd.DataFrame(summary_data)
print("\nSummary Table - Quantum Model:")
print(summary_df.to_string(index=False))

# Classical comparison summary
if classical_results:
    print("\n" + "="*80)
    print("COMPARISON RESULTS (Classical vs Quantum)")
    print("="*80)
    
    classical_summary = []
    for results in classical_results:
        classical_summary.append({
            "Seed": results["seed"],
            "Best Val Acc": f"{results['best_val_accuracy']:.4f}",
            "F1-Score": f"{results['f1_score']:.4f}",
        })
    
    classical_df = pd.DataFrame(classical_summary)
    print("\nClassical Model Results (No Quantum):")
    print(classical_df.to_string(index=False))
    
    # Compare averages
    quantum_avg_acc = np.mean([r['best_val_accuracy'] for r in all_results])
    classical_avg_acc = np.mean([r['best_val_accuracy'] for r in classical_results])
    
    print(f"\n{'='*50}")
    print("PERFORMANCE COMPARISON:")
    print(f"Quantum Model Avg Accuracy: {quantum_avg_acc:.4f}")
    print(f"Classical Model Avg Accuracy: {classical_avg_acc:.4f}")
    print(f"Improvement with Quantum: {(quantum_avg_acc - classical_avg_acc)*100:.2f}%")
    print(f"{'='*50}")

# ========================== STATISTICS ACROSS SEEDS ==========================

metrics = ["best_val_accuracy", "final_val_accuracy", "final_train_accuracy",
           "f1_score", "precision", "recall"]

stats = {}
for metric in metrics:
    values = [r[metric] for r in all_results]
    stats[metric] = {
        "mean": np.mean(values),
        "std":  np.std(values),
        "min":  np.min(values),
        "max":  np.max(values),
    }

stats_df = pd.DataFrame(stats).round(4)
print("\n" + "="*80)
print("STATISTICS ACROSS 5 SEEDS (Quantum Model)")
print("="*80)
print(stats_df)
```
